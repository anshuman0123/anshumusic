```python
from surprise import Dataset,SVD,Reader
```

```python
new_df['anime_id'].unique()
```

```python
new_df['name'].unique()
```

```python
anime_name = dict()
index = 0
for x,y in zip(new_df['anime_id'].unique(),new_df['name'].unique()):
    anime_name[index]=y
    index+=1
index=0
anime_name_new = dict()
for x,y in anime_name.items():
    anime_name_new[y]=x
```

```python
new_df1 = new_df.drop('name',axis=1)
new_df1
```

```python
reader = Reader(rating_scale=(1,10))
data = Dataset.load_from_df(new_df1,reader=reader)
```

```python
trainset = data.build_full_trainset()
```

```python
# algo = SVD()
# similarity_matrix = algo.fit(trainset).compute_similarities()
```

```python
from surprise import KNNBasic
sim_options = {'name': 'cosine',
               'user_based': False  # compute  similarities between items
               }
algo = KNNBasic(sim_options=sim_options)
similarity_matrix = algo.fit(trainset).compute_similarities()
```

```python
similarity_matrix.shape
```

```python
similarity_matrix
```

```python
def recommend(name):
    suggestions = []
    index = anime_name_new[name]
    similar_items = sorted(list(enumerate(similarity_matrix[index])),key=lambda x:x[1],reverse=True)[:6]
    for i in similar_items:
        suggestions.append(anime_name[i[0]])
    return suggestions
```

```python
recommend('Highschool of the Dead')
```

```python
from surprise.dump import dump,load
```

```python
dump('algo.h5',predictions=similarity_matrix,algo=KNNBasic,verbose=1)
```
